# 딕셔너리(dictionary)

```python
# dict 자료 구조 생성

## 데이터가 없는 빈 껍질 생성
dict_data = {}
print(dict_data)

## key=문자열, value=문자열
dict_data2 = {'name':'리쿠'}
print(dict_data2)

## key = 숫자, value = 문자열
dict_data3 = {1 : '리쿠' , 2 : '유우시'}
print(dict_data3)

## key = 문자열, value = list
dict_data4 = {'name' : ['리쿠','유우시'], 'job' : ['아이돌' , '댄서']}
print(dict_data4)
```

```python
# 딕셔너리 생성 연습 문제
custom_info = {'name' : 'HM' , 'age' :20 , 'phone' : '010-1234-1234'}
print(custom_info)
```

```python
# 딕셔너리 값(value) 추출

## 고객 이름 추출
name = custom_info['name']
print(name)

## 고객 나이 추출
age = custom_info['age']
print(age)

## 고객의 연락처 추출
number = custom_info['phone']
print(number)
```

```python
# 데이터 추가 및 수정

## 새로운 key : value 쌍 추가
custom_info['birth'] = '06/28'
print(f'데이터 추가 후 결과 확인 : {custom_info}')

## 데이터 수정:기존의 key:value -> key에 대한 value 수정
custom_info['birth'] = '06/14'
print(f'데이터 수정 후 결과 확인 : {custom_info}')
```

```python
# 데이터 추가 실습

## 변수 dict_test를 다음과 같이 만들기
dict_test = {'노래제목' : '아무노래'}
print(dict_test)

## 데이터 추가를 통해 다음과 같이 정보를 저장하기
dict_test['가수'] = '지코'
print(dict_test)

dict_test['날짜'] = '2020.01.13'
print(dict_test)
```

```python
# 보충 : 존재하지 않는 key -> 값 추출

## 방법 1 : 변수['key']
data = dict_data['email']
#print(data)

print('-'*80)

## 방법 2 : 변수.get('key')
data = dict_data.get('email')
print(data)

print('-'*80)
```

# 딕셔너리 연습 문제

```python
# 문제 1: 나만의 인물 사전 만들기
## 좋아하는 연예인의 이름(name)과 직업(job)을 key로 하는 딕셔너리 my_star를 만들고,딕셔너리 전체를 출력하는 코드를 작성하세요

my_star = {'name' : '리쿠', 'job' : '아이돌'}
print(my_star)
```

```python
# 문제 2: 딕셔너리 값 추출하여 문장 만들기
## 나의 도시(city)와 인구(population) 정보가 담긴 my_city 딕셔너리가 있습니다. 각 key를 이용해 값을 추출하고, f-string을 사용하여 "제가 사는 곳은 OOO이고, 인구는 OO만 명입니다." 형식으로 출력하는 코드를 작성하세요.

my_city = {'city' : '광주' , 'population' : 40}

## 딕셔너리 값 추출 -> 변수['key']
location = my_city['city']
people = my_city['population']
print(f'제가 사는 곳은 {location}이고, 인구는 {people}만 명입니다.')
```

```python
# 문제 3: 새로운 정보 추가하기
## 상품의 이름과 가격이 저장된 product 딕셔너리에, key가 'stock'이고 value가 100인 재고 정보를 추가한 후, 정보가 추가된 product 딕셔너리 전체를 출력하는 코드를 작성하세요.


product = {'name': '커피', 'price': 5000}
product['stock'] = 100
print(product)
```

```python
# 문제 4: 기존 정보 수정하기
## 문제 3의 product 딕셔너리에서, 가격('price')이 5000원에서 5500원으로 인상되었습니다. 가격 정보를 수정하고, 수정된 product 딕셔너리 전체를 출력하는 코드를 작성하세요.

product['price'] = 5500
print(product)
```

```python
# 문제 5: 리스트를 값으로 갖는 딕셔너리 만들기
## 영화 제목(title)과 출연 배우 목록(actors)을 key로 하는 딕셔너리 movie를 만드세요. 출연배우는 2명 이상으로 하여 리스트 자료구조를 value로 사용해야 합니다. 딕셔너리를 만든 후 전체를 출력하는 코드를 작성하세요.

movie = {'title' : '왕과 사는 남자', 'actors' : ['박지훈','유해진']}
print(movie)
```

## 내장 함수

### 리스트 전용 함수 - append()

```python
# 기존 데이터에 값 추가 -> append() 함수

## 정수 0부터 4까지 채워진 리스트 데이터 생성
list_data = [0,1,2,3,4]
print(f'데이터 추가 전 리스트 자료형 확인: {list_data}')

## 마지막 위치에 5 추가
list_data.append(5) # 혹은 [0,1,2,3,4].append(5)도 가능
print(f'데이터 추가 후 결과 확인: {list_data}')
```

```python
# 기존 데이터 삭제 예약어 del

## del 변수[인덱스]
del list_data[5]
print(f'데이터 삭제 후 결과 확인 : \n{list_data}')
```

```python
# 성분 원소의 개수 확인 -> 범용 함수: len() 함수

## 리스트 자료형 생성
list_data = [1,2,3,4,5,6,7,8,9]

## len(변수) -> 성분 원소의 개수 추출
number = len(list_data)
print(f'리스트 자료형 데이터 안에 들어있는 성분 원소의 개수 = {number}개')
```

### 딕셔너리 전용 함수 - keys()

```python
# key만 가져오기(추출)

## 딕셔너리 자료 구조 데이터 생성
dict_data = {'name':'리쿠', 'age':24, 'phone':'010-1234-5567'}
print(dict_data)

## 전체 key 추출 -> keys() 함수 호출
keys = dict_data.keys()
print(f'추출된 전체 key 확인 : {keys}')

## keys() 함수 호출의 결과 값 -> 자료형 확인 -> type() 함수 호출
print(f'추출된 전체 key의 자료형 확인 : {type(keys)}')
```

```python
# value 가져오기(추출)

## 딕셔너리 자료 구조 데이터 -> dict_data
print(dict_data)

## 전체 value 추출 -> values() 함수 호출
values = dict_data.values()
print(f'추출된 전체 value 확인 : {values}')

## values() 함수 호출의 결과 값 -> 자료형 확인 -> type() 함수 호출
print(f'추출된 전체 value의 자료형 확인 : {type(values)}')
```

```python
# key로 value 가져오기(추출)

## 딕셔너리 자료 구조 데이터 -> dict_data
print(dict_data)

## name의 값 추출
#name = dict_data['name']
name = dict_data.get('name')
print(f'name의 값 추출 : {name}')
```

## 함수 연습문제

```python
# 문제1 장바구니에 물건 담고 개수 세기
## 빈 리스트로 cart 변수를 만든 후, append()를 사용하여 '계란', '우유', '사과'를 차례대로 추가하세요. 마지막으로 len()을 사용하여 장바구니에 총 몇 개의 물품이 담겼는지 출력하는 코드를 작성하세요.

cart = []
cart.append('계란')
cart.append('우유')
cart.append('사과')
count = len(cart)
print(f'장바구니에 담긴 물품의 개수: {count}개')
```

```python
# 문제2 내 정보의 항목 확인하기
## 이름, 도시, 이메일 주소를 포함하는 my_info 딕셔너리를 만드세요. 그 후 keys() 함수를 사용하여 이 딕셔너리가 어떤 정보 항목(key)들을 가지고 있는지 확인하고 출력하는 코드를 작성하세요.

my_info = {'name':'리쿠', 'city': 'seoul', 'email': 'skdldbwjd@naver.com'}
keys = my_info.keys()
print(f'내 정보 항목: {keys}')
```

```python
# 문제3 시험 점수만 모아보기
## 국어, 영어, 수학 과목의 점수가 저장된 scores 딕셔너리가 있습니다. values() 함수를 사용하여 모든 과목의 점수들만 추출하여 출력하는 코드를 작성하세요.

scores = {'국어': 95, '영어': 88, '수학': 100}
values = scores.values()
print(values)
```

```python
# 문제4 get()으로 안전하게 정보 조회하기
## 책의 제목(title)과 저자(author) 정보가 담긴 book 딕셔너리가 있습니다. get() 함수를 사용하여 이 책의 저자를 writer 변수에 저장하고 출력하는 코드를 작성하세요.

book = {'title': 'AI 에이전트 심화 프로젝트:LangGraph와 MCP로 완성하는 업무 자동화', 'author': '이영배'}
writer = book.get('author')
print(writer)
```
